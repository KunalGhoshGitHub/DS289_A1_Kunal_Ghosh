```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Output_files = pd.read_csv("Output_file_names.csv",delimiter=",",header=None).to_numpy()
Output_files = np.squeeze(Output_files)
Output_files = Output_files.tolist()
```

```python
Input_File = open("Values_of_step_size.txt", "r")
Step_sizes = []
for line in Input_File:
    if line[0] != "/" and (not line[0].isspace()):
        Step_sizes.append(line)
Input_File.close() 
```

```python
Files_Corresponding_to_Step_sizes = []
for step_size in Step_sizes:
    for file_name in Output_files:
        if step_size.strip() in file_name:
            Files_Corresponding_to_Step_sizes.append(file_name)
            print(file_name)
    print("***********************")
```

```python
Files_Corresponding_to_Step_sizes = []
for step_size in Step_sizes:
    Files_Corresponding_to_Step_sizes = []
    for file_name in Output_files:
        if step_size.strip() in file_name:
            Files_Corresponding_to_Step_sizes.append(file_name)
            #print(file_name)
    Analytical_Solution_grid_points = pd.read_csv(Files_Corresponding_to_Step_sizes[0], delimiter = ",",header=None).to_numpy()
    Analytical_Solution = pd.read_csv(Files_Corresponding_to_Step_sizes[1], delimiter = ",",header=None).to_numpy()
    Euler_Explicit_Solution_grid_points = pd.read_csv(Files_Corresponding_to_Step_sizes[2], delimiter = ",",header=None).to_numpy()
    Euler_Explicit_Solution = pd.read_csv(Files_Corresponding_to_Step_sizes[3], delimiter = ",",header=None).to_numpy()
    Adam_Solution_grid_points = pd.read_csv(Files_Corresponding_to_Step_sizes[4], delimiter = ",",header=None).to_numpy()
    Adam_Solution = pd.read_csv(Files_Corresponding_to_Step_sizes[5], delimiter = ",",header=None).to_numpy()
    RK4_Solution_grid_points = pd.read_csv(Files_Corresponding_to_Step_sizes[6], delimiter = ",",header=None).to_numpy()
    RK4_Solution = pd.read_csv(Files_Corresponding_to_Step_sizes[7], delimiter = ",",header=None).to_numpy()
    
    fig=plt.figure()
    plt.plot(Analytical_Solution_grid_points,Analytical_Solution)
    plt.plot(Euler_Explicit_Solution_grid_points,Euler_Explicit_Solution)
    plt.plot(Adam_Solution_grid_points,Adam_Solution)
    plt.plot(RK4_Solution_grid_points,RK4_Solution)
    plt.legend(["Analytical", "Euler Explicit", "Adams-Bashforth", "RK4"])
    plt.title("Step size: "+step_size.strip())
    plt.xlabel("t")
    plt.ylabel("y")
    plt.grid()
    plt.show()
    Files_Corresponding_to_Step_sizes = []
    for file_name in Output_files:    
        fig.savefig("Q_2_b_Solution_"+ str(step_size)+"_.png",dpi = 500)
```

```python
fig=plt.figure()
Files_Corresponding_to_Step_sizes = []
Legend = []
for file_name in Output_files: 
    if "Euler_Explicit_Solution" in file_name:
        for step_size in Step_sizes:
            if step_size.strip() in file_name:
                Files_Corresponding_to_Step_sizes.append(file_name)
                if len(Files_Corresponding_to_Step_sizes) == 2:
                    Euler_Explicit_Solution_grid_points = pd.read_csv(Files_Corresponding_to_Step_sizes[0], delimiter = ",",header=None).to_numpy()
                    Euler_Explicit_Solution = pd.read_csv(Files_Corresponding_to_Step_sizes[1], delimiter = ",",header=None).to_numpy()
                    
                    plt.plot(Euler_Explicit_Solution_grid_points,Euler_Explicit_Solution)
                    Files_Corresponding_to_Step_sizes = []
            Legend.append("Step size = "+step_size.strip())
plt.legend(Legend,bbox_to_anchor=(1, 1))
plt.title("Solution using Euler’s explicit method")
plt.xlabel("t")
plt.ylabel("y")
plt.grid()
plt.show()
fig.savefig("Q_2_b_Euler_Explicit_Solution.png",dpi = 500,bbox_inches="tight")
```

```python
fig=plt.figure()
Files_Corresponding_to_Step_sizes = []
for file_name in Output_files: 
    if "Adam_Solution" in file_name:
        for step_size in Step_sizes:
            if step_size.strip() in file_name:
                Files_Corresponding_to_Step_sizes.append(file_name)
                if len(Files_Corresponding_to_Step_sizes) == 2:
                    Adam_Solution_grid_points = pd.read_csv(Files_Corresponding_to_Step_sizes[0], delimiter = ",",header=None).to_numpy()
                    Adam_Solution = pd.read_csv(Files_Corresponding_to_Step_sizes[1], delimiter = ",",header=None).to_numpy()
                    
                    plt.plot(Adam_Solution_grid_points,Adam_Solution)
                    Files_Corresponding_to_Step_sizes = []
            Legend.append("Step size = "+step_size.strip())
plt.legend(Legend,bbox_to_anchor=(1, 1))
plt.title("Solution using Second-order Adams-Bashforth Method")
plt.xlabel("t")
plt.ylabel("y")
plt.grid()
plt.show()
fig.savefig("Q_2_b_Adam_Solution.png",dpi = 500,bbox_inches="tight")
```

```python
fig=plt.figure()
Files_Corresponding_to_Step_sizes = []
for file_name in Output_files: 
    if "RK4_Solution" in file_name:
        for step_size in Step_sizes:
            if step_size.strip() in file_name:
                Files_Corresponding_to_Step_sizes.append(file_name)
                if len(Files_Corresponding_to_Step_sizes) == 2:
                    RK4_Solution_grid_points = pd.read_csv(Files_Corresponding_to_Step_sizes[0], delimiter = ",",header=None).to_numpy()
                    RK4_Solution = pd.read_csv(Files_Corresponding_to_Step_sizes[1], delimiter = ",",header=None).to_numpy()
                    
                    plt.plot(RK4_Solution_grid_points,RK4_Solution)
                    Files_Corresponding_to_Step_sizes = []
            Legend.append("Step size = "+step_size.strip())
plt.legend(Legend,bbox_to_anchor=(1, 1))
plt.title("Solution using Fourth-order RK method Method")
plt.xlabel("t")
plt.ylabel("y")
plt.grid()
plt.show()
fig.savefig("Q_2_b_RK4_Solution.png",dpi = 500,bbox_inches="tight")
```

```python
Average_error_Euler_Explicit = []
for file_name in Output_files:
    if "Average_error" in file_name:
        if "Euler_Explicit" in file_name:
            print(file_name)
            Average_error_Euler_Explicit = pd.read_csv(file_name, delimiter = ",",header=None).to_numpy().squeeze()
```
