```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Output_files = pd.read_csv("Output_file_names.csv",delimiter=",",header=None).to_numpy()
Output_files = np.squeeze(Output_files)
Output_files = Output_files.tolist()
```

```python
Input_File = open("Input_Values_of_c.txt", "r")
C_s = []
for line in Input_File:
    if line[0] != "/" and (not line[0].isspace()):
        C_s.append(line)
Input_File.close() 
```

```python
for file_name in Output_files:
    if "Grid_Points" in file_name:
        Grid_Points = pd.read_csv(file_name, delimiter = ",",header=None).to_numpy()
```
