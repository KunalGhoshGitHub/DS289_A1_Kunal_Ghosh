# Import the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Reading the name of the files generated

```python
Output_files = pd.read_csv("Output_file_names.csv",delimiter=",",header=None).to_numpy()
Output_files = np.squeeze(Output_files)
Output_files = Output_files.tolist()
```

# Reading the file corresponding to the different values of c

```python
Input_File = open("Input_Values_of_c.txt", "r")
C_s = []
for line in Input_File:
    if line[0] != "/" and (not line[0].isspace()):
        C_s.append(line)
Input_File.close() 
```

# Reading the file corresponding to grid points

```python
for file_name in Output_files:
    if "Grid_Points" in file_name:
        Grid_Points = pd.read_csv(file_name, delimiter = ",",header=None).to_numpy()
```

# Plotting for different values of c

```python
Files_Corresponding_to_c_s = []
Legend = []
fig = plt.figure()
for c in C_s:
    Files_Corresponding_to_Step_sizes = []
    for file_name in Output_files:
        if c.strip() in file_name:
            if "_1_" in file_name:
                #Files_Corresponding_to_c_s.append(file_name)
                #print(file_name)
                Legend.append("c = "+c.strip())
                x = pd.read_csv(file_name, delimiter = ",",header=None).to_numpy()    
                plt.plot(Grid_Points,x)
plt.legend(Legend)
#plt.title("Displacement versus time for each of the three values of the damping coefficient")
plt.xlabel("t (in s)")
plt.ylabel("x (in m)")
plt.grid()
plt.show()
fig.savefig("Q_3_Displacement_versus_time_for_different_c.png",dpi = 500,bbox_inches = 'tight')
```

# Converting the values of step sizes for to a list

```python
dt = pd.read_csv("Values_of_dt.csv",delimiter=",",header=None).to_numpy()
dt = np.squeeze(dt)
dt = dt.tolist()
```

# Creating legends for the graph

```python
Legend_dt = []
for d_t in dt:
    a = r"$\Delta t = "+str(d_t)+"s $"
    Legend_dt.append(a)
```

# Plotting for displacement vs time for c = 5 and different step sizes

```python
i = 0;
fig = plt.figure()
Legend=[]
for file_name in Output_files:
    if "RK4_1Solution" in file_name:
        x = pd.read_csv(file_name, delimiter = ",",header=None).to_numpy()
        Grid_Points = np.linspace(0,15,num=len(x))
        plt.plot(Grid_Points,x)
        i = i+1
plt.xlabel(r"$t\:(in\:s)$")
plt.ylabel(r"$x\:(in\:m)$")
plt.legend(Legend_dt,bbox_to_anchor=(1, 1))
plt.grid()
i = i+1;
plt.show()
fig.savefig("Q_3_Convergence_of_results_for_reducing_step_size.png",dpi = 500,bbox_inches = 'tight')
```
