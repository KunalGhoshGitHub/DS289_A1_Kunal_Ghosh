# Import the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Reading the name of the files generated

```python
Output_files = pd.read_csv("Output_file_names.csv",delimiter=",",header=None).to_numpy()
Output_files = np.squeeze(Output_files)
Output_files = Output_files.tolist()
```

# Reading the file corresponding to the different values of S

```python
Input_File = open("Input_Values_of_S.txt", "r")
S_s = []
for line in Input_File:
    if line[0] != "/" and (not line[0].isspace()):
        S_s.append(line)
Input_File.close() 
```

# Plotting the variation of temperature along the radial direction for different values of S

```python
Files_Corresponding_to_S = []
for S in S_s:
    Files_Corresponding_to_S = []
    for file_name in Output_files:
        if "_"+str(float(S.strip())) in file_name:
            Files_Corresponding_to_S.append(file_name)
    r = pd.read_csv(Files_Corresponding_to_S[0], delimiter = ",",header=None).to_numpy()
    T = pd.read_csv(Files_Corresponding_to_S[1], delimiter = ",",header=None).to_numpy()
    fig = plt.figure()
    plt.plot(r,T)
    plt.title("S ="+S.strip())
    plt.xlabel("r")
    plt.ylabel("T")
    plt.grid()
    plt.show()
    fig.savefig("Q_5_S_is_"+S.strip()+".png",dpi = 400,bbox_inches = "tight")
```

# Plotting the variation of temperature along the radial direction for different values of S in the same graph for comparison

```python
Files_Corresponding_to_S = []
Legend = []
fig = plt.figure()
for S in S_s:
    Files_Corresponding_to_S = []
    for file_name in Output_files:
        if "_"+str(float(S.strip())) in file_name:
            Files_Corresponding_to_S.append(file_name)
    r = pd.read_csv(Files_Corresponding_to_S[0], delimiter = ",",header=None).to_numpy()
    T = pd.read_csv(Files_Corresponding_to_S[1], delimiter = ",",header=None).to_numpy()
    Legend.append("S = "+S.strip())
    plt.plot(r,T)
plt.xlabel("r")
plt.ylabel("T")
plt.legend(Legend)
plt.grid()
plt.show()
fig.savefig("Q_5_Different_Values_of_S.png",dpi = 400,bbox_inches = "tight")
```
